## Analysis and results

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# For numerical columns: Correlation matrix
numerical_cols = ['followers', 'engage_others', 'favorites', 'neuroticism', 'tweet_per_day']
correlation_matrix = df[numerical_cols + ['extraversion']].corr()
print(correlation_matrix['extraversion'])

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix with Extraversion")
plt.show()

# For object columns: Graph them against extraversion
object_cols = ['tweet_length', 'use_hashtags', 'include_links', 'composition_difference', 'common_topics', 'changes_over_time', 'accuracy_of_data', 'reflect_personality', 'engage_with_similar_traits', 'geographic_influence', 'linguistic_features', 'emotional_support', 'importance_of_twitter']

for col in object_cols:
    plt.figure(figsize=(10, 6))
    sns.barplot(x=col, y='extraversion', data=df)
    plt.title(f"{col} vs. Extraversion")
    plt.ylabel("Average Extraversion")
    plt.show()
```

```python
import pandas as pd
from google.colab import files
import chardet

# Upload the CSV file
uploaded = files.upload()

# Assuming you uploaded only one file, extract its name
file_name = list(uploaded.keys())[0]

# Detect the encoding of the CSV file
with open(file_name, 'rb') as file:
    result = chardet.detect(file.read())

encoding = result['encoding']

# Now, use the detected encoding to read the CSV
df = pd.read_csv(file_name, encoding=encoding)

# Display the DataFrame
print(df.columns)
```

```python
columns_to_drop = ['id_str', 'name', 'screen_name','location','description','listed_count','number_of_quoted_statuses','number_of_tweets_scraped']
df = df.drop(columns=columns_to_drop)
print(df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Descriptive Statistics for each MBTI type
print(df.groupby('mbti_personality').describe().transpose())

# Boxplots for each numeric column vs. mbti_personality
numeric_columns = ['followers_count', 'friends_count', 'favourites_count', 'statuses_count', 'number_of_retweeted_statuses', 'total_retweet_count', 'total_favorite_count', 'total_hashtag_count', 'total_url_count', 'total_mentions_count', 'total_media_count', 'average_tweet_length', 'average_retweet_count', 'average_favorite_count', 'average_hashtag_count', 'average_url_count', 'average_mentions_count', 'average_media_count']

for col in numeric_columns:
    plt.figure(figsize=(12, 6))
    sns.boxplot(x='mbti_personality', y=col, data=df)
    plt.title(f"Boxplot of {col} vs. MBTI Personality")
    plt.xticks(rotation=45)
    plt.show()

# Countplot for mbti_personality
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='mbti_personality', order=df['mbti_personality'].value_counts().index)
plt.title("Distribution of MBTI Personality")
plt.xticks(rotation=45)
plt.show()

```

## Conclusions

###Survey Data Findings:

1.   The survey data analysis shows that people who believe they are extraverted have a overall higher twitter usage along with a higher number of followers, tweets per day, engagement with others.
2.   On the other hand, Neuroticism seemed to have little to no correlation with high twitter use.

###Twitter MBTI Data Findings:

Analysing approximately 8000 users' twitter activity including their followers count, friends count, retweet length along with another parameter, there is no clear trend established between the 16 MBTI personalities and twitter use.

###Several factors could account for the divergence between the survey data results and the Twitter MBTI data:

**Scope and Nature of Data Collection**:
The methodology and context in which the survey data and Twitter data were collected might be inherently different. While the survey might have been a structured questionnaire capturing people's perceptions at a specific point in time, the Twitter data represents more dynamic, ongoing user behaviors.

**Sample Representativeness**:
The survey might have been administered to a specific subset of the population, perhaps restricted by age, geography, or other demographics, leading to results that don't generalize well to the broader Twitter user base. The MBTI data from Twitter, on the other hand, may span a wider and more diverse user demographic, introducing variability.

**Subjectivity vs. Objectivity**:
The survey data hinges on individuals' self-assessment, which can be influenced by cognitive biases, societal expectations, or even the desire to project a specific image. In contrast, the MBTI data from Twitter represents tangible metrics and actions, which tend to be more objective.

**Personal Bias and Dissonance Between Belief and Behavior**:
One's MBTI designation may not always align with one's self-perception and behavior. While MBTI provides a standardized classification based on certain criteria, personal biases can influence how one perceives and reports their own traits. For instance, someone might identify as an extravert based on MBTI but still feel introverted in certain social settings, leading to discrepancies between their self-reported behavior and actual online activity.



###Inference from Results

The exploration of both survey data and Twitter MBTI data provides a compelling look into the intricate relationship between perceived personality traits and online behavior. The survey data indicates a noticeable trend: those identifying as extraverted tend to have a higher affinity for Twitter, marked by more followers, more frequent tweeting, and increased engagement. Interestingly, neuroticism did not showcase a similar trend. This contrasts starkly with the Twitter MBTI data, which, even after examining the behaviors of roughly 8000 users, fails to present a clear correlation between any of the 16 MBTI personalities and Twitter usage patterns.

This divergence underscores several critical points. Firstly, the methodological differences between the two data sources cannot be ignored. Surveys capture static self-perceptions at a single time point, while Twitter data provides a more fluid, ongoing record of user behavior. Secondly, the audience demographics for each data set may differ, leading to variations in the derived insights. Surveys, depending on their distribution, might not capture as diverse a demographic as the broader Twitter community.

A significant point of contention emerges from the subjectivity inherent in self-reported survey data. Respondents' self-perceptions can be clouded by cognitive biases, societal norms, or even personal aspirations, making their self-reports potentially less reliable. This contrasts with the objectivity of metrics-driven Twitter data, where actions, rather than perceptions, are recorded.

Moreover, the intriguing issue of personal bias and the potential dissonance between belief and behavior arises. It's conceivable for someone's MBTI designation, a relatively standardized measure, to clash with their self-perception, a discrepancy that can manifest in both survey responses and online behavior.

###Future Developments:

Unified Methodology: Future research should take into account harmonizing approaches, making sure that self-report surveys and behavior-driven data sources are coordinated in their data gathering methodology.

Broader Demographics: Ensuring that the sample group reflects the Twitter community's broader demographics can produce results that are easier to extrapolate.

Longitudnal Studies: Studies that follow the same group of people over an extended period of time might offer insights into changing attitudes and actions, offering light on the dynamic interaction between personality and online behavior.

## Word Count

The following code will count the number of words in Markdown cells. Code cells are not included.

- `Main word count` is the number of words in the main body of the text, *excluding* references or appendices.
- `References and appendices word count` is the number of words in any references or appendices.

Only `Main word count` relates to the assignment word limit. There is no limit to the number of words that can be included in references or appendices. Please note that appendices should only be used to provide context or supporting information. *No marks will be directly awarded for material submitted in appendices*.

Important:

- Please do not modify the word count code!
- To exclude references from your word count **you must** have a cell that starts with the text `## References`. Everything below this cell will not count towards the main word count.
- If you are submitting additional material as appendices **you must** have a cell that starts with the text `## Appendices`. Everything below this cell will not count towards the main word count. If you do not have any appendices you can delete the `## Appendices` cell.
- Code comments should only be used to explain details of the implementation, not for discussing your findings. All analysis commentary **must** be written in Markdown cells. *No marks will be awarded for analysis discussion submitted as comments in code cells*.

```python
%%js

// Run this cell to update your word count.

function wordcount() {
    let wordCount = 0
    let extraCount = 0
    let mainBody = true

    let cells = Jupyter.notebook.get_cells()
    cells.forEach((cell) => {
        if (cell.cell_type == 'markdown') {
            let text = cell.get_text()
            // Stop counting as main body when get to References or Appendices.
            if (text.startsWith('## References') ||
                text.startsWith('## Appendices')) {
                mainBody = false
            }
            if (text.startsWith('## Word Count')) {
                text = ''
            }
            if (text) {
                let words = text.toLowerCase().match(/\b[a-z\d]+\b/g)
                if (words) {
                    let cellCount = words.length
                    if (mainBody) {
                        wordCount += cellCount
                    } else {
                        extraCount += cellCount
                    }
                }
            }
        }
    })
    return [wordCount, extraCount]
}

let wc = wordcount()
element.append(`Main word count: ${wc[0]} (References and appendices word count: ${wc[1]})`)
```
